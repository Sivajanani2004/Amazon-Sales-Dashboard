### Cleaning the Datas and Handling the missing values

```python
Amazon_Sales.columns = Amazon_Sales.columns.str.strip().str.lower().str.replace(' ', '_')  # standardize names
```

```python
Amazon_Sales.isnull().sum()
```

```python
Cleaned_data = Amazon_Sales.dropna(inplace = True)
```

```python
Cleaned_data
```

```python
Amazon_Sales.isnull().sum()
```

```python
Amazon_Sales["amount"] = pd.to_numeric(Amazon_Sales["amount"], errors='coerce')
Amazon_Sales["qty"] = pd.to_numeric(Amazon_Sales["qty"], errors='coerce')
```

```python
Amazon_Sales["Total_sale"] = Amazon_Sales["qty"] * Amazon_Sales["amount"]
Amazon_Sales["Order_count"] = 1
```

```python
Amazon_Sales["ship-state"] = Amazon_Sales["ship-state"].str.title()
Amazon_Sales["category"] = Amazon_Sales["category"].str.upper()
```

### Exporting the Cleaned data

```python
Amazon_Sales.to_csv("Cleaned_amazon_sales.csv",index = False)
```
